```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df_students=pd.read_csv('StudentsPerformance.csv')
# print(df_students.head())
print("Используется dataset 'Успеваемость студентов на экзаменах.\n\
Оценки, полученные студентами по различным предметам'\n\
источник - https://www.kaggle.com/spscientist/students-performance-in-exams")
```

```python
sns_plot = sns.histplot(df_students['math score'])
plt.grid(True)
fig = sns_plot.get_figure()
plt.title('Диаграмма распределения оценок по математике')
plt.xlabel('Оценка')
plt.ylabel('Частота')
plt.show()
print("Вывод: по математике сутденты болше хороших оценок, чаще всего получают от 60 до 80 баллов.")
sns_plot = sns.histplot(df_students['reading score'])
plt.grid(True)
fig = sns_plot.get_figure()
plt.title('Распределение оценок по чтению')
plt.xlabel('Оценка')
plt.ylabel('Частота')
plt.show()
print("Вывод: по чтению сутденты чаще встречаются оценки от 55 до 85, при этом в отличии от математики \n\
чаще встречются очень плохие оценки.")
sns_plot = sns.histplot(df_students['writing score'])
plt.grid(True)
fig = sns_plot.get_figure()
plt.title('Распределение оценок по письму')
plt.xlabel('Оценка')
plt.ylabel('Частота')
plt.show()
print("Вывод: по письму сутденты чаще встречаются оценки от 50 до 85, как и по чтению \n\.")
```

```python
df_students['mean_score']= df_students.apply(lambda row: round((row['math score']+row['reading score']+row['writing score'])/3,2), axis=1)

sns.boxplot(x="test preparation course", y="mean_score", data=df_students)
print("Вывод: на диаграграмме явно видно, что успеваемость студентов, посещавщих подготовительные курсы выше. \n\
Выше и сама средняя оценка и гораздо меньше плохих оценок.\n\
---Для исходных данных посчитана средняя оценка по трем предметам и построена диаграграмма boxplot")
```

```python
pivot = df_students.pivot_table(
    index=["parental level of education"],
    columns=["test preparation course"],
    values="mean_score",
    aggfunc=np.average)
sns.heatmap(pivot)
print("Вывод: на тепловой карте видно, что успеваемость студентов, посещавщих подготовительные курсы \n\
и имеющих родителей с учеными степенями выше. \n\
---Для исходных данных посчитана средняя оценка по трем предметам и построена диаграграмма HeatMap")
```

```python
ax = sns.regplot(data=df_students, x='reading score', y='writing score', color='green', marker='+')
ax.set_title('Корреляция между оценками по чтению и письму')
print("Вывод: На диаграмме видна прямая связь между оценками по чтению и письму\n\
---диаграграмма распредения")
```
