## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'A222BC96'`

Результат: 
`Номер A222BС валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
import re
car_id = 'А222ВС96'

def car_valid(car_id):
    x=re.search(r'[АAВBЕEКKМMНHОOРPСCТTУYХX]\d{3}[АAВBЕEКKМMНHОOРPСCТTУYХX]{2}\d{2}',car_id)
    if x != None:
        return ('Номер '+re.search(r'\w\d{3}\w{2}',x.group()).group()+' валиден. Регион: '+re.search(r'\d{2}$',x.group()).group())
    else:
        return (f'Номер не валиден')
# print(re.search(r'\d{2}$',x.group()).group())
# print(re.search(r'\w\d{3}\w{2}',x.group()).group())
car_valid('А222BB93')
````

## Задание 3

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:  

1. `some_words = 'Информационные технологии'`

Результат: 
`ИТ`

2. `some_words = 'Near Field Communication'`

Результат: 
`NFC`

````python
some_words = 'Информационные технологии'
some_words = 'Near Field Communication'
def upper_case(str_):
    return str_[0][0].upper()
def to_acronym(str_):
    return re.sub(r'(\s)','',(re.sub(r'(\w+)',upper_case, str_)))
print(to_acronym(some_words))
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com','ddd','sd@mail.ru']
domains={}
domain_pattern=r'(?<=@)\w+.\w+'
for email in emails:
    domain=re.findall(domain_pattern, email)
    if domain:
        if domains.get(domain[0])!=None:
              domains[domain[0]]+=1
        else:
            domains[domain[0]]=1
for domain,count in domains.items():
    print(f'{domain}: {count}')
        
````

+7-xxx-xxx-xx-xx  
Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера.
Примеры работы программы:  

1. `phone = '+7 955 555-55-55'`

`+7-950-555-55-55`

2. `phone = '8(955)555-55-55'`
