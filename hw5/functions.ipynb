````python
 documents =  [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
# перечень полок, на которых хранятся документы ( если документ есть в documents, то он обязательно должен быть и в 
directories = {
'1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def shelf_count_docs( doc_shelf,l_directories ):
    s = [len( docs ) for directory , docs  in l_directories.items(  ) if directory == doc_shelf]
    if len( s ) > 0:
        return s[0]
    else:
        return None
    
def list_of_documents( l_documents,l_directories ):
    print( 'Текущий список документов:' )
    for document in l_documents:
        doc_shelf = [directory for directory , docs  in l_directories.items(  ) if document['number'] in docs][0]
        print( f'№: {document["number"]}, тип: {document["type"]}, владелец: {document["name"]}, полка хранения: {doc_shelf}' )
        
def doc_by_number( doc_number,l_documents ):
    s = [i for i,document  in enumerate( l_documents ) if document['number'] == doc_number]
    if len( s ) > 0:
        
        return s[0]
    else:
        print( 'Документ не найден в базе.' )
        return None
    
def list_of_shelf( l_directories ):
    print( f'Текущий перечень полок: {( ", " ).join( l_directories.keys(  ) )}' )

def shelf_of_document( doc_number,l_directories ):
    s = [directory for directory , docs  in l_directories.items(  ) if doc_number in docs]
    if len( s ) > 0:
        return s[0]
    else:
        return None
    
def find_owner_by_doc_number( l_documents ):
    doc_number = input( "Введите номер документа: " )
    num_doc = doc_by_number( doc_number,l_documents )  
    print( f'Владелец документа: {l_documents[num_doc]["name"]}' ) if num_doc != None else None
    

def find_shelf_by_doc_number( l_directories ):
    doc_number = input( "Введите номер документа: " )
    shelf = shelf_of_document( doc_number,l_directories )
    print( f'Документ хранится на полке: {shelf}' ) if shelf != None else print( 'Документ не найден в базе' )

def create_new_shelf( l_directories ):
    doc_shelf = input( "Введите номер полки: " )
    if shelf_count_docs( doc_shelf,l_directories ) == None:
        l_directories[doc_shelf] = []
        print( f'Полка добавлена.' )
        list_of_shelf( l_directories )
    else:
        print( f'Полка уже существует.' )
        list_of_shelf( l_directories )

def drop_shelf( l_directories ):
    doc_shelf = input( "Введите номер полки: " )
    if shelf_count_docs( doc_shelf,l_directories ) == None:
         s_prn = 'Такой полки не существует.'
    elif shelf_count_docs( doc_shelf,l_directories ) == 0:
        del l_directories[doc_shelf]
        s_prn = 'Полка удалена.'
    else:
        s_prn = 'На полке есть документы, удалите их перед удалением полки.'
    print( f'{s_prn}' )
    list_of_shelf( l_directories )

def create_new_doc( l_documents, l_directories ):
    doc_number = input( 'Введите номер документа: ' )
    doc_type = input( 'Введите тип документа: ' )
    doc_owner = input( 'Введите владельца документа: ' )
    doc_shelf = input( 'Введите полку для хранения:' )
    if shelf_count_docs( doc_shelf,l_directories ) != None:
        l_documents.append( {'type': doc_type, 'number': doc_number, 'name': doc_owner} )
        l_directories[doc_shelf].append( doc_number )
    else:    
        print( 'Такой полки не существует. Добавьте полку командой ads.' )
    list_of_documents( l_documents,l_directories )
            
def drop_doc( l_documents, l_directories ):
    doc_number = input( "Введите номер документа: " )
    if doc_by_number( doc_number,l_documents ) != None:
        l_documents.pop( doc_by_number( doc_number,l_documents ) )
        l_directories[shelf_of_document( doc_number,l_directories )].remove( doc_number )
        print( 'Документ удален.' )        
    list_of_documents( l_documents,l_directories )
            
def moving_doc( l_documents,l_directories ):
    doc_number = input( 'Введите номер документа: ' )
    doc_shelf = input( 'Введите номер полки: ' )
    if doc_by_number( doc_number,l_documents ) != None:
        if shelf_count_docs( doc_shelf,l_directories ) == None:
            print( f'Такой полки не существует.' )
            list_of_shelf( l_directories )
        else:
            l_directories[shelf_of_document( doc_number,l_directories )].remove( doc_number )
            l_directories[doc_shelf].append( doc_number )
            print( 'Документ перемещен.' )
            list_of_documents( l_documents,l_directories )

            
def main( l_documents, l_directories ): 
    while True:
        command_name = input( 'Введите команду: ' )
        print('')
        if command_name == 'ad':
            create_new_doc( l_documents, l_directories )
            
        elif command_name == 'd':
            drop_doc( l_documents, l_directories )

        elif command_name == 'm':
            moving_doc( l_documents,l_directories )
            
        elif command_name == 'p':
            find_owner_by_doc_number( l_documents )
            
        elif command_name == 's':
            find_shelf_by_doc_number( l_directories )
        
        elif command_name == 'l':
            list_of_documents( l_documents,l_directories )
            
        elif command_name == 'ads':
            create_new_shelf( l_directories )
           
        elif command_name == 'ds':
            drop_shelf( l_directories )
           
     
        elif command_name == 'q':
            break
            
main( documents,directories )
````
